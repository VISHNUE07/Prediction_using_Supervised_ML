## **Problem Statement**
---

Predict the percentage of a student based on the number of study hours.

### Importing Libraries

```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

### Importing Data

```python
url = 'https://raw.githubusercontent.com/AdiPersonalWorks/Random/master/student_scores%20-%20student_scores.csv'
df = pd.read_csv(url)
df.head()
```

```python
# summary of a dataframe

df.info()
```

```python
# statistical summary of the dataframe

df.describe()
```

```python
# data visualisation

sns.scatterplot(x='Hours', y='Scores', data=df)
```

### Dataset Splitting

```python
from sklearn.model_selection import train_test_split
X_train , X_test , y_train , y_test = train_test_split( df['Hours'], df['Scores'] , test_size = 0.2 , random_state = 82)
```

```python
# visualising training and testing data

sns.scatterplot(x=X_train, y=y_train, label='Training Data')
sns.scatterplot(x=X_test, y=y_test, label='Testing Data')
plt.title('Model Visualisation')
```

### Model Training

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train.values.reshape(-1,1), y_train)
```

### Model Testing

```python
# predicting values

y_pred = model.predict(X_test.values.reshape(-1,1))
predicted = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
predicted
```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)


print('Mean Absolute Error : ',mae)
print('Mean Squared Error : ',mse)
print('Root Mean Squared Error : ',rmse)
print('R2 Score : ',r2)
```

```python
# Plotting on Test Data

plt.plot(X_test , y_pred , label = 'LinearRegression' , color = 'green')
plt.scatter (X_test , y_test , label = 'Test_data' , color = 'purple'  )
plt.legend()
plt.show()
```

### Testing with new value

```python
x = model.predict([[9.25]])
print("No. of Hours = 9.25")
print("Predicted Score = {}".format(x[0]))
```
